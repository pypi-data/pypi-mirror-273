```python
# Visual Python: Data Analysis > Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Visual Python: Data Analysis > File
vp_df = pd.read_csv('https://raw.githubusercontent.com/visualpython/visualpython/main/visualpython/data/sample_csv/iris.csv')
vp_df
```

```python
# Visual Python: Data Analysis > Frame
vp_df = vp_df.copy()
vp_df['variety'] = vp_df['variety'].shift(1)
vp_df.set_index(['variety'], inplace=True)
vp_df.reset_index(inplace=True)
vp_df['variety'] = vp_df['variety'].shift(2)
vp_df['variety'] = vp_df['variety'].fillna(vp_df['variety'].mode()[0])
vp_df
```
