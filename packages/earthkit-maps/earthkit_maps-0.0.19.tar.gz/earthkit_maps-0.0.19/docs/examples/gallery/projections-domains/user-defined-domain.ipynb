## User-defined domain

```python
import earthkit.maps
from earthkit.maps.domains import Domain

AZORES = Domain(bounds=[-32, -24, 36.5, 40], domain_name="Azores")

chart = earthkit.maps.Chart(domain=AZORES)

chart.land()
chart.cities(density="high")
chart.ocean()
chart.coastlines()
chart.gridlines()

chart.subplot_titles("{domain}")

chart.show()
```
